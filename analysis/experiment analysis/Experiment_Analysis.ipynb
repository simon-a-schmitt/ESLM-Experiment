```python
# Create the DataFrame
df = pd.DataFrame(data)

df.head(5)
```

```python
df["KGE"] = df["Model"].apply(lambda x: "KGE" in x)
```

```python
df['Model base'] = df['Model'].str.split(' ').str[0]
```

```python
df.head(20)
```

```python
columns = ["Training Time Total", "Training Time Mean", "Prediction Time Single Triples"]
rows = ["Top 5", "Top 10", "BERT", "ERNIE", "T5", "KGE", "w/o KGE"]
df_aggr = pd.DataFrame(index=rows, columns=columns)
```

```python
# Group by Top 5 vs top 10
# Look at differences in computational requirements

df_top5 = df[df["Topk"] == 5]
df_top10 = df[df["Topk"] == 10]

print("Training Time Total")
print("Top 5", df_top5["Training Time Total"].mean())
df_aggr.loc["Top 5", "Training Time Total"] = df_top5["Training Time Total"].mean()
print("Top 10", df_top10["Training Time Total"].mean())
df_aggr.loc["Top 10", "Training Time Total"] = df_top10["Training Time Total"].mean()
print("Training Time Mean")
print("Top 5", df_top5["Training Time Mean"].mean())
df_aggr.loc["Top 5", "Training Time Mean"] = df_top5["Training Time Mean"].mean()
print("Top 10", df_top10["Training Time Mean"].mean())
df_aggr.loc["Top 10", "Training Time Mean"] = df_top10["Training Time Mean"].mean()
print("Prediction Time Single Triples")
print("Top 5", df_top5["Prediction Time Single Triples"].mean())
df_aggr.loc["Top 5", "Prediction Time Single Triples"] = df_top5["Prediction Time Single Triples"].mean()
print("Top 10", df_top10["Prediction Time Single Triples"].mean())
df_aggr.loc["Top 10", "Prediction Time Single Triples"] = df_top10["Prediction Time Single Triples"].mean()

# A: No significat differences
```

```python
# Group by model base (BERT, ERNIE, T5)
# Look at differences in computational requirements

df_bert = df[df["Model base"] == "BERT"]
df_ernie = df[df["Model base"] == "ERNIE"]  
df_t5 = df[df["Model base"] == "T5"]    

print("Training Time Total")
print("BERT", df_bert["Training Time Total"].mean())
df_aggr.loc["BERT", "Training Time Total"] = df_bert["Training Time Total"].mean()
print("ERNIE", df_ernie["Training Time Total"].mean())
df_aggr.loc["ERNIE", "Training Time Total"] = df_ernie["Training Time Total"].mean()
print("T5", df_t5["Training Time Total"].mean())
df_aggr.loc["T5", "Training Time Total"] = df_t5["Training Time Total"].mean()
print("Training Time Mean")
print("BERT", df_bert["Training Time Mean"].mean())
df_aggr.loc["BERT", "Training Time Mean"] = df_bert["Training Time Mean"].mean()
print("ERNIE", df_ernie["Training Time Mean"].mean())
df_aggr.loc["ERNIE", "Training Time Mean"] = df_ernie["Training Time Mean"].mean()
print("T5", df_t5["Training Time Mean"].mean())
df_aggr.loc["T5", "Training Time Mean"] = df_t5["Training Time Mean"].mean()
print("Prediction Time Single Triples")
print("BERT", df_bert["Prediction Time Single Triples"].mean())
df_aggr.loc["BERT", "Prediction Time Single Triples"] = df_bert["Prediction Time Single Triples"].mean()
print("ERNIE", df_ernie["Prediction Time Single Triples"].mean())
df_aggr.loc["ERNIE", "Prediction Time Single Triples"] = df_ernie["Prediction Time Single Triples"].mean()
print("T5", df_t5["Prediction Time Single Triples"].mean())
df_aggr.loc["T5", "Prediction Time Single Triples"] = df_t5["Prediction Time Single Triples"].mean()

# A: T5 is the most computationally expensive model
```

```python
# Group by with/without KGE
# Look at differences in computational requirements

df_kge = df[df["KGE"] == True]
df_no_kge = df[df["KGE"] == False]

print("Training Time Total")
print("KGE", df_kge["Training Time Total"].mean())
df_aggr.loc["KGE", "Training Time Total"] = df_kge["Training Time Total"].mean()
print("No KGE", df_no_kge["Training Time Total"].mean())
df_aggr.loc["w/o KGE", "Training Time Total"] = df_no_kge["Training Time Total"].mean()
print("Training Time Mean")
print("KGE", df_kge["Training Time Mean"].mean())
df_aggr.loc["KGE", "Training Time Mean"] = df_kge["Training Time Mean"].mean()
print("No KGE", df_no_kge["Training Time Mean"].mean())
df_aggr.loc["w/o KGE", "Training Time Mean"] = df_no_kge["Training Time Mean"].mean()
print("Prediction Time Single Triples")
print("KGE", df_kge["Prediction Time Single Triples"].mean())
df_aggr.loc["KGE", "Prediction Time Single Triples"] = df_kge["Prediction Time Single Triples"].mean()
print("No KGE", df_no_kge["Prediction Time Single Triples"].mean())
df_aggr.loc["w/o KGE", "Prediction Time Single Triples"] = df_no_kge["Prediction Time Single Triples"].mean()

# A: Integration of KGE increases computational requirements for training and inference
```

```python
# Group by dataset
# Look at differences in computational requirements

df_dbpedia = df[df["Dataset"] == "DBpedia"]
df_linkedmdb = df[df["Dataset"] == "LinkedMDB"]
df_faces = df[df["Dataset"] == "FACES"]

print("Training Time Total")
print("DBpedia", df_dbpedia["Training Time Total"].mean())
print("LinkedMDB", df_linkedmdb["Training Time Total"].mean())
print("FACES", df_faces["Training Time Total"].mean())
print("Training Time Mean")
print("DBpedia", df_dbpedia["Training Time Mean"].mean())
print("LinkedMDB", df_linkedmdb["Training Time Mean"].mean())
print("FACES", df_faces["Training Time Mean"].mean())
print("Prediction Time Single Triples")
print("DBpedia", df_dbpedia["Prediction Time Single Triples"].mean())
print("LinkedMDB", df_linkedmdb["Prediction Time Single Triples"].mean())
print("FACES", df_faces["Prediction Time Single Triples"].mean())
```
